```python
def findSum(x,y):
  sum=0
  for i in range(x,y+1):
    sum+=i
  return sum

print('sum is',findSum(1,10))
```

```python
class Calculation:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def findSum(self):
        sum = 0
        for i in range(self.x, self.y + 1):
            sum += i
        return sum

    def average(self):
        sum = self.findSum()
        average = sum / (self.y - self.x + 1)
        return average

# Example usage
calc = Calculation(1, 5)
print("Sum:", calc.findSum())     # Output: Sum: 15
print("Average:", calc.average()) # Output: Average: 3.0
```

#List, Turple, Set

```python
# Define a list
fruits = ["apple", "banana", "cherry", "orange"]
fruits.insert(0,"watermelon")
print(fruits)
print(fruits[0:3])


print(len(fruits))
print(type(fruits))
print(dir(fruits)) #try out remove and append
```

```python
# Creating a list
my_list = [1, 2, 3, 4, 5]

# Modifying elements
my_list[0] = 10

print(my_list)

# Appending elements
my_list.append(6)


print(my_list)

# Removing elements
my_list.remove(3)

print(my_list)
```

```python
# Creating a tuple
my_tuple = (1, 2, 3, 4, 5)
x=my_tuple[0]
# Accessing elements
print(x)  # Output: 1

# Tuples are immutable - this would raise an error:
# my_tuple[0] = 10
```

# Combine List with ForLoop

```python
# Define a list
fruits = ["apple", "banana", "cherry", "orange"]
print(fruits)
print(fruits[0:3])

# Iterate over each item in the list and print them
for fruit in fruits:
    print(fruit)
print(type(fruits))
print(dir(fruits))
```

```python
# Iterating over a tuple
my_tuple = (1, 2, 3, 4, 5)
for item in my_tuple:
    print(item)

# Iterating over a set
my_set = {1, 2, 3, 4, 5}
for item in my_set:
    print(item)
```
